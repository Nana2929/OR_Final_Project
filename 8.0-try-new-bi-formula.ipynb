```python
N_NODE = 16
def make_temp_bi(df, 
                n_node = N_NODE):
    def get_bi(bike_s, bike_d, 
               pole_s, pole_d):
        bike_ = bike_s - bike_d
        pole_ = pole_s - pole_d 
        print(bike_, pole_, end = ' -> ')
        res = bike_
        # 缺車又缺柱，有人會騎來，有人想騎走，假設是可以馬上換手的
        if bike_ < 0 and pole_ < 0:
            res =  -(pole_) + bike_ 
        # 先滿足bike不足的數量
        elif bike_ < 0:
            if abs(bike_) > pole_s:
                bike_ = - pole_s
                res = bike_ 
        # 再考慮無車柱可還，需移車才能空出車位ㄉ狀態
        elif pole_ < 0:  
            res = abs(pole_) 
        print(res)
        return res
    
    for time in ['早九', '晚五']:
        df[f'{time}_tempbi'] = df.apply(lambda x: 
                                       get_bi(x[f'{time}可借車數_pre'], x[f'{time}借車人數'], x[f'{time}空柱數量_pre'], x[f'{time}還車人數']), axis = 1)
        print('----')

    # df.drop(df.index[[1,3]])
```

```python
import joblib
def gen_cases(data, 
              dmatrixpath = './data/mini-testcases/mini_dmat.txt'):
    dmat = np.loadtxt(dmatrixpath, delimiter=',')
    for time in ['早九', '晚五']:
        dire = './data/mini-testcases'
        file_prefix = f'{dire}/mini_newb'
        b = data[f'{time}_tempbi'].tolist()
        b.insert(0, -sum(b))
        testcase = {'n_node':N_NODE, 'b':b, 'd':dmat} 
        joblib.dump(testcase, f'{file_prefix}_{time}')
gen_cases(minidf)
```
